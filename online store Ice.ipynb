# Test the following hypotheses

### Average user ratings of the Xbox One and PC platforms are the same

- H0: Average user ratings of the Xbox One and PC platforms are the same.
- H1: Average user ratings of the Xbox One and PC platforms are different.

<div class="alert alert-success" role="alert">
Reviewer's comment v. 1:
    
Good job that you formulated the null and alternative hypotheses explicitly.
</div>

```python
xone_user_ratings = new_game_df.query('platform == "XOne" & user_score > 0')
pc_user_ratings = new_game_df.query('platform == "PC" & user_score > 0')

alpha = 0.05 

print('Xbox One user score variance', xone_user_ratings['user_score'].var())
print('PC user score variance', pc_user_ratings['user_score'].var())

results = st.ttest_ind(xone_user_ratings['user_score'], pc_user_ratings['user_score'], equal_var=False)

print('p-value: ', results.pvalue)

if (results.pvalue < alpha):
        print("we reject the null hypothesis")
else:
        print("we can't reject the null hypothesis")
```

With a p-value of 0.05, we can reject the null hypothesis and say that the platforms have different user ratings.

### Average user ratings for the Action and Sports genres are different

- H0: Average user ratings for the Action and Sports genres are the same.
- H1: Average user ratings for the Action and Sports genres are different.

<div class="alert alert-success" role="alert">
Reviewer's comment v. 2:
    
Yes, hypotheses formulation is correct :)
</div>

```python
action_user_rating = new_game_df.query('genre == "Action" & user_score > 0')
sports_user_rating = new_game_df.query('genre == "Sports" & user_score > 0')

alpha = 0.05 

print('Action user score variance', action_user_rating['user_score'].var())
print('Sports user score variance', sports_user_rating['user_score'].var())

results = st.ttest_ind(action_user_rating['user_score'], sports_user_rating['user_score'])

print('p-value: ', results.pvalue)

if (results.pvalue < alpha):
        print("we reject the null hypothesis")
else:
        print("we can't reject the null hypothesis")
```

With a p-value of 0.05, we can reject the null hypothesis and say that the Average user ratings for the Action and Sports genres are different
